```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

def scrape_all_books():
    """
    Raspa todos os livros do site https://books.toscrape.com
    Retorna um DataFrame com as colunas: nome, preco, url
    Salva o arquivo itens.xlsx no diretório atual.
    """
    base_url = "https://books.toscrape.com"
    all_books = []   # lista de dicionários

    page = 1
    while True:
        list_url = f"{base_url}/catalogue/page-{page}.html"
        print(f"Processando página {page}...")

        try:
            response = requests.get(list_url, timeout=10)
            response.raise_for_status()
        except requests.exceptions.RequestException as e:
            print(f"Erro ao acessar {list_url}: {e}")
            break

        soup = BeautifulSoup(response.text, 'lxml')

        # Encontra todos os livros da página
        books = soup.find_all('article', class_='product_pod')
        if not books:
            print("Nenhum livro encontrado, encerrando.")
            break

        for book in books:
            # Título e link
            title_tag = book.find('h3').find('a')
            title = title_tag.get('title', '').strip()
            href = title_tag.get('href')

            # Preço (remove símbolo da moeda e troca ponto por vírgula)
            price_tag = book.find('p', class_='price_color')
            if price_tag:
                price_raw = price_tag.text.strip().replace('£', '').replace('€', '')
                price_formatted = price_raw.replace('.', ',')
            else:
                price_formatted = ''

            # URL completa
            if href.startswith('catalogue/'):
                full_url = f"{base_url}/{href}"
            else:
                full_url = f"{base_url}/catalogue/{href}"

            all_books.append({
                'nome': title,
                'preco': price_formatted,
                'url': full_url
            })

            # Pequena pausa para não sobrecarregar o servidor
            time.sleep(0.2)

        # Verifica se existe a próxima página
        next_li = soup.find('li', class_='next')
        if not next_li:
            break
        page += 1

    # Cria o DataFrame
    df = pd.DataFrame(all_books)

    # Exporta para Excel (nome exato conforme solicitado)
    df.to_excel('itens.xlsx', index=False)
    print(f"Arquivo 'itens.xlsx' salvo com {len(df)} livros.")
    return df

if __name__ == "__main__":
    scrape_all_books()
```
